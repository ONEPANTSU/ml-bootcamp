# Introduction to NumPy
___
<b>NumPy</b> is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

### 4.4. Dot product

```python
np.random.seed(0)

mat1 = np.random.randint(10, size=(5, 3))
mat2 = np.random.randint(10, size=(5, 3))

mat1, mat2
```

```python
# Element-wise multiplication (Hadamard product)
mat1 * mat2
```

```python
# Dot product
mat1 @ mat2.T # np.dot(mat1, mat2)
```

### <i>Dot product example (Nut butter sales)</i>

```python
# Create sales_amounts array

np.random.seed(0)

sales_amounts = np.random.randint(20, size=(5, 3))
sales_amounts                                  
```

```python
# Create weekly_sales DataFrame

weekly_sales = pd.DataFrame(sales_amounts, 
                            index=["Mon", "Tues", "Wed", "Thurs", "Fri"],
                            columns=["Almond butter", "Peanut butter", "Cashew butter"])
weekly_sales
```

```python
# Create prices array

prices = np.array([10, 8, 12])
prices
```

```python
prices.shape
```

```python
# Create butter_prices DataFrame

butter_prices = pd.DataFrame(prices.reshape(1, 3),
                            index=["Price"],
                            columns=["Almond butter", "Peanut butter", "Cashew butter"])
butter_prices
```

```python
# Compute total_sales araray

total_sales = prices @ sales_amounts.T
total_sales
```

```python
# Create daily_sales DataFrame

daily_sales = butter_prices @ weekly_sales.T
daily_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
